```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('mcmc/COM_PowerSpect_CMB-TT-full_R3.01.txt')
```

```python
np.shape(data)
```

```python
l = data[:,0]
var = data[:,1]
varerr = 1/2*(data[:,2]+data[:,3])
```

```python
fig, ax = plt.subplots()
ax.plot(l,var,'.')
ax.errorbar(l,var,varerr,color = 'k',alpha = 0.5)
ax.set_xlabel(r'Multipole $$')
fig.tight_layout()
```
